```python
import matplotlib.pyplot as plt
import numpy as np
import collections
import csv
import os
```

```python
file_name = 'mqtt2'
summary_file_path = 'benchmarks/working_dir/' + file_name + '.csv'
execution_times = []
with open(summary_file_path, 'r') as inFile:
    read = csv.reader(inFile, delimiter=',')
    
    for row in read:
        execution_times.append(row)

executions = [int(item[0]) for item in execution_times]
values = [int(item[1]) for item in execution_times]

average = sum(values)/len(values)
std_dev = np.std(values)

plt.plot(executions, values, 'o-')
plt.hlines(average, 0.0, 100, 'g', linestyle='-')


plt.title('Runtime Benchmark - ' + str(file_name), fontsize=15)
plt.xlabel('Execution Index')
plt.ylabel('Time | µs')


print("Standard Deviation: " + str(std_dev))
print("Coefficient of Variation: " + str((std_dev/average)*100))
print("Average: " + str(average))
print("")
print("")
```
